## #02. Series 객체 생성

### 1. list를 사용한 Series 객체 생성

- Series는 기존의 연속성 자료형을 가공하여 생성된 데이터 구조
- list나 ndarray는 인덱스가 암묵적으로 존재하지만 Series는 인덱스가 명시적으로 존재한다.

```python
# 리스트를 통한 객체 생성
my_list = [10,30]
s1 = Series(my_list)
s1
```

### 2. ndarray 배열을 사용한 Series 객체 생성

```python
# ndarray를 통한 객체 생성
my_array = np.array([100, 200])
s2 = Series(my_array)
s2
```

### 3. 딕셔너리를 통한 객체 생성

- 딕셔너리의 key가 index 역할을 한다.

```python
# 딕셔너리를 통한 객체 생성
my_dict = {"hello": 123, "world": 456, "nice": 789}
s3 = Series(my_dict)
s3
```

## #04. 인덱싱과 슬라이싱

### 1. list나 ndarray를 통해 생성한 경우

- 일반적인 연속성 자료형의 경우와 동일

```python
my_series = Series([10, 20, 30, 40, 50])
print("== 인덱싱 ==")
print(my_series[0])
print("== 슬라이싱 ==")
print(my_series[1:3])
```

### 2. 딕셔너리를 통해 생성한 경우

- 인덱스가 문자열로 존재한다.

```python
my_series = Series({"a": 100, "b": 200, "c": 300, "d": 400})
print("== 인덱싱 ==")
print(my_series["a"])
print("== 슬라이싱 ==")
print(my_series["b":"d"])
```

## #05. Series의 값만 추출하기

### 1. `values` 프로퍼티를 사용해서 값만 추출

- ndarray 형식임을 알 수 있다.

```python
my_series = Series({"a": 100, "b": 200, "c": 300, "d": 400})

print(type(my_series.values))
my_series.values
```

### 2. 값을 list 자료형으로 변환

```python
mylist = list(my_series.values)
mylist
```

## #06. Series의 인덱스만 추출하기

### 1. `index` 프로퍼티 사용

```python
my_series = Series({"a": 100, "b": 200, "c": 300, "d": 400})

print(type(my_series.index))
my_series.index
```

### 2. 인덱스를 list 타입으로 변환

```python
mylist = list(my_series.index)
mylist
```
