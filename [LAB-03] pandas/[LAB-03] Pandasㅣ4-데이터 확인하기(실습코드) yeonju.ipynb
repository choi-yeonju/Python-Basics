## #08. 기술 통계량 확인 

### 1. 컬럼 단위 확인 
- DataFrame 의 컬럼은 Series 객체 이므로 Series의 메서드를 그대로 사용 가능함 

```python
print("합계:", origin['국어'].sum())
print("평균:", origin['국어'].mean())
print("최대:", origin['국어'].max())
print("최소:", origin['국어'].min())
print("분산:", origin['국어'].var())
print("표준편차:", origin['국어'].std())
print("중앙값:", origin['국어'].median())
print("1사분위 수:", origin['국어'].quantile(0.75))
print("2사분위 수:", origin['국어'].quantile(0.25))
print("3사분위 수:", origin['국어'].quantile(0.75))
```

### 2. 명목형 변수의 기술 통계량 
- 학년의 경우 데이터 타입이 INT이기 때문에 집계되지 않는다. 
    - str, object, category 타입만 집계됨 
- count : 총 도수 
- unique : 값의 계급(종류) 수 
- top : 최빈값

### 1. 학년을 포함한 명목형 기술 통계량 집계 
- 학년 컬럼의 데이터 타입을 명목형으로 변경 후 \

```python
origin['학년'] = origin['학년'].astype('category')
origin.describe(include= ['object', 'category']).T
```
