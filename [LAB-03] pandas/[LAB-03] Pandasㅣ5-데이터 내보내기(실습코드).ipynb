## #01. 준비작업

```python
from jussam import load_data
```

### 2. 데이터 불러오기

- 과목별 점수 데이터

```python
origin = load_data('grade')
origin
```

## #02. 엑셀로 저장하기

### 1. 파일 이름만 지정

```python
origin.to_excel("1.xlsx")
```

###  2. 파라미터 설정 - 인덱스를 포함하는 경우

```python
origin.to_excel("2.xlsx",           # 파일이름
                index=True,         # 인덱스 포함 여부 (기본값=True)
                index_label="Name"  # 인덱스 필드의 제목
        )
```

### 3. 파라미터 설정 - 인덱스를 포함하지 않는 경우

```python
origin.to_excel("3.xlsx",                       # 파일이름
                index=False,                    # 인덱스 포함 여부
                columns=['학년', '성별', '국어'], # 저장할 컬럼 지정
                header=['level', 'sex', 'kor']  # 컬럼별 제목 설정
            )
```

## #03. CSV로 내보내기

### 1. 원본 그대로 저장하기

- encoding을 설정하는 것 외에는 엑셀파일을 내보내는 것과 동일하다.

```python
origin.to_csv("1.csv", encoding='utf-8')
```

###  2. 파라미터 설정 - 인덱스를 포함하는 경우

```python
origin.to_csv("2.csv",          # 파일이름
            encoding="utf-8",   # 파일 인코딩
            index=True,         # 인덱스 포함 여부 (기본값=True)
            index_label="Name"  # 인덱스 필드의 제목
        )
```

### 3. 파라미터 설정 - 인덱스를 포함하지 않으면서 컬럼 이름을 직접 설정

```python
origin.to_csv("3.csv",                      # 파일이름
            encoding="utf-8",               # 인코딩
            index=False,                    # 인덱스 포함 여부
            columns=['학년', '성별', '국어'], # 저장할 컬럼 지정
            header=['level', 'sex', 'kor']  # 컬럼별 제목 설정
        )
```
