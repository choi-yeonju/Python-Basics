# [LAB-03] 2. Series 객체의 기능

## 연습문제

```python
import pandas as pd

# 학생의 시험 점수 딕셔너리
scores = {"국어": 85, "영어": 92, "수학": 78, "과학": 95, "사회": 88}

# 1. 딕셔너리를 Series 객체로 변환
score_series = pd.Series(scores)
print("--- 전체 점수 ---")
print(score_series)

# 2. 수학 과목의 점수 출력
math_score = score_series["수학"]
print(f"--- 수학 점수 ---")
print(f"수학 점수: {math_score}")

# 3. 90점 이상인 과목 선택
high_scores = score_series[score_series >= 90]
print("--- 90점 이상 과목 ---")
print(high_scores)

# 4. 모든 과목 점수의 평균 계산
average_score = score_series.mean()
print(f"--- 평균 점수 ---")
print(f"평균 점수: {average_score}")
```
