### #02. Series 객체 생성(1)

### 1. list를 통한 Series 객체 생성 
- Series는 기존의 연속성 자료형을 가공하여 생성된 데이터 구조
- list나 naarray는 인덱스 번호가 암묵적으로 존재하지만 Series는 인덱스가 명시적으로 존재한다.

```python
# 리스트를 통한 객체 생성 
my_list = [10, 30]
s1 = Series(my_list)
s1
```

```python
my_list2 =["영","#$@$#"]
s2 = Series(my_list2)
s2
```

### 2. ndarray를 통한 Series 객체 생성 

```python
# ndarray를 통한 Series 객체 생성 
my_array = np.array([100, 200])
s2 = Series(my_array)
s2
```

```python
s2 = Series(np.array([100, 200]))
s2
```

```python
# 이론적으로는 이것도 됨
import pandas as 바나나
s = 바나나.Series([1, 2, 3])  # 작동은 함
s
```

### 3. 딕셔너리를 통한 생성 
- 딕셔너리의 key가 index 역할을 한다.

```python
# 딕셔너리를 통한 객체 생성 
my_dict = {"Hello":123, "world":456, "nice":789}
s3 = Series(my_dict)
s3
```

## #03. Series의 정보 확인 

### 1. 데이터 타입 확인 

```python
my_series = Series({"hello": 123, "world": 456, "nice":789})
my_series.dtype
```

### 2. 종합적인 정보 확인

```python
my_series.info()
```

```python
# <class 'pandas.Series'>
# → 이 객체는 pandas Series야.

# Index: 3 entries, hello to nice
# → 인덱스가 문자열이고, hello부터 nice까지 총 3개 항목.

# Series name: None
# → 이 Series에 이름이 없음. (pd.Series([...], name="가격") 이렇게 주면 채워짐)

# Non-Null Count  Dtype
# 3 non-null      int64
# → 3개 중 결측값(NaN) 없이 전부 값 있음. 타입은 int64.

# dtypes: int64(1)
# → int64 타입 컬럼이 1개. (DataFrame에서 더 유용한 정보, Series는 항상 1)

# memory usage: 48.0+ bytes
# → 이 Series가 메모리 48바이트 차지. +는 문자열 인덱스 때문에 추가 메모리 있다는 뜻.
```

## #04. 인덱싱과 슬라이싱 
### 1. list나 ndarray를 통해 생성한 경우 

```python
my_series = Series([10, 20, 30, 40, 50])
print("==인덱싱==")
print(my_series[0])
print("==슬라이싱==")
print(my_series[1:3])
```

```python
### 2. 딕셔너리를 통해 생성한 경우 

my_series = Series({"a":100, "b":200,"c":300,"d":400})
print("==인덱싱==")
print(my_series["a"])
print("==슬라이싱==")
print(my_series["b":"d"])
```

## #05. Series의 값만 추출하기 

### 1. values 프로퍼티를 사용해서 값만 추출 

```python
my_series = Series({"a":100,"b":200,"c":300,"d":400})

print(type(my_series.values))
my_series.values
```

### 2. 값을 list 자료형으로 변환

```python
my_list = list(my_series.values)
my_list
```

### #06. Series의 인덱스만 추출하기 

### 1. index 프로퍼티 사용

```python
my_series = Series({"a":100, "b":200, "c":300, "d":400})

print(type(my_series.index))
my_series.index
```

```python
### 2. 인덱스를 list 타입으로 변환 

mylist = list(my_series.index)
mylist
```
