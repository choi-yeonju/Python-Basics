## #02. Series 객체의 연산

### 1. 실험을 위한 객체 만들기

```python
# 리스트나 ndarray를 통한 객체 생성시 인덱스를 직접 지정하면서 만들기
my1 = Series([10, 20, 30], index=['hello', 'world', 'nice'])
my1
```

```python
# 딕셔너리로 생성
my2 = Series({"world": 1, "test": 2, "hello": 3})
my2
```

### 3. Serise 객체에 대한 숫자 연산

모든 원소에 대해 연산을 수행한다. (NumPy도 동일)

```python
addon = my1 + 1000
addon
```

```python
times = my1 * 5
times
```

## #03. 기술 통계량

### 1. ndarray의 기능을 그대로 지원

```python
column = Series([188, 92, 45, 63, 79])

print("합계:", column.sum())
print("평균:", column.mean())
print("최대:", column.max())
print("최소:", column.min())
print("분산:", column.var())
print("표준편차:", column.std())
print("중앙값:", column.median())
print("1사분위 수:", column.quantile(0.25))
print("2사분위 수:", column.quantile(0.5))
print("3사분위 수:", column.quantile(0.75))
```

## #04. 조건검색

### 1. 샘플 데이터 정의

- 0부터 100 전까지 (99까지) 10씩 증가하는 값을 원소로 갖는 객체 생성

```python
column = Series(np.arange(0, 100, 10))
column
```

### 2. Numpy와 동일한 표현식 사용

```python
# numpy와 같은 표현식을 사용한다.
in1 = column[column > 30]
in1
```

### 3. `AND` 검색

- 대괄호를 이중으로 표기한다.

```python
in2 = column[column <= 70][column > 10]
in2
```

### 4. `OR` 검색

- 하나의 대괄호 안에 파이프라인(`|`)으로 구분한다.

```python
in3 = column[(column <= 10) | (column >= 70)]
in3
```

## #05. 타입변환

### 1. 실습을 위한 원본 배열 구성

```python
origin = Series(np.arange(0, 100, 10))
origin.info()
```

### 2. 실수(`float`)타입으로 변환

Numpy와 마찬가지로 `astype()`메서드를 사용한다.

```python
x = origin.astype('float')
x.info()
```

### 3. 문자열(`str`)타입으로 변환

```python
y = origin.astype('str')
y.info()
```

## #06. 데이터 정렬

### 1. 값 기반 정렬 (오름차순)

```python
my_dict = {"c": 456, "a": 123, "b": 789}
origin = Series(my_dict)

sort1 = origin.sort_values()
sort1
```

### 2. 값 기반 정렬 (내림차순)

```python
sort2 = origin.sort_values(ascending=False)
sort2
```

### 3. 인덱스 기반 정렬 (인덱스 오름차순)

```python
sort3 = origin.sort_index()
sort3
```

### 4. 인덱스 기반 정렬 (인덱스 내림차순)

```python
sort4 = origin.sort_index(ascending=False)
sort4
```
