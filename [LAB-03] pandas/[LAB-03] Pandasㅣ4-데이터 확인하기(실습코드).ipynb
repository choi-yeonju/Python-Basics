## #06. 값의 변경

## #08. 기술 통계량 확인

### 1. 컬럼 단위 확인

- DataFrame의 각 컬럼은 Series 객체이므로 Series의 메서드를 그대로 사용 가능함

```python
print("합계:", origin['국어'].sum())
print("평균:", origin['국어'].mean())
print("최대:", origin['국어'].max())
print("최소:", origin['국어'].min())
print("분산:", origin['국어'].var())
print("표준편차:", origin['국어'].std())
print("중앙값:", origin['국어'].median())
print("1사분위 수:", origin['국어'].quantile(0.25))
print("2사분위 수:", origin['국어'].quantile(0.5))
print("3사분위 수:", origin['국어'].quantile(0.75))
```

### 2. DataFrame 단위 확인

- 모든 연속형 컬럼에 대해서 일괄 계산한다.
- 계산할 수 없는 형태의 값도 계산되므로 주의가 필요하다.

```python
origin.sum()
```

### 3. 컬럼 단위로 모든 기술 통계량 일괄 산출

```python
origin['국어'].describe()
```

### 4. DataFrame 단위로 일괄 산출

- 연산이 가능한 연속형 변수만 추출해서 적용해야 한다.

```python
# 연산이 가능한 컬럼만 추출
df2 = origin[['국어','영어','수학','과학']]
desc = df2.describe()
desc
```

### 5. 전치 구하기

- 가로와 세로가 바뀐 새로운 데이터프레임을 생성할 수 있다.

```python
t = desc.T
t
```

## #09. 명목형 변수의 기술 통계량 확인

### 1. 빈도수 구하기 → 값의 종류별 수

- 예제 데이터에서 "성별", "학년" 등과 같이 연산을 수행할 수 없는 카테고리 성격의 데이터는 종류별 수량을 구해서 데이터 분포를 파악한다.

```python
origin['성별'].value_counts()
```

```python
origin['학년'].value_counts()
```

### 2. 명목형 변수 기술 통계량

- 학년의 경우 데이터 타입이 int이기 때문에 집계되지 않는다.
  - str, object, category 타입만 집계됨
- **count**: 총 도수
- **unique** : 값의 계급(종류) 수
- **top**: 최빈값
- **freq** : 최빈값의 데이터 수

```python
origin.describe(include='object').T
```

### 3. 학년을 포함한 명목형 기술 통계량 집계

- 학년 컬럼의 데이터 타입을 명목형으로 변경 후   
  기술통계량 집계 대상을 object 타입과 category 타입으로 지정한다.

```python
origin['학년'] = origin['학년'].astype('category')
origin.describe(include=['object', 'category']).T
```
