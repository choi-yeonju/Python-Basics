### 라이브러리 참조

```python
from jussam import load_data
from pandas import Series
```

###  데이터 가져오기

```python
monsters = load_data('monsters')
monsters
```

### 1. 데이터의 첫인상

```python
monsters.info()
```

### 2. 쓸모 없는 정보 삭제

```python
df1 = monsters.drop(columns='last_seen')
df1.head()
```

### 3. 신규 개체 추가

```python
new_monster = Series({'height': 2.2, 'species': 'Kelpie', 'caution_level': 'Medium'})

df2 = df1.copy()
df2.loc['M021'] = new_monster
df2
```

### 3. 데이터 오류 수정

```python
# 인덱스도 M003_copy로 변경해야 하므로 인덱스 해제
df3 = df2.reset_index()

df3.loc[3] = df3.loc[2].copy()
df3.loc[3, 'monster_id'] = 'M003_copy'
df3
```

### 5. 최종 보고서용 정렬

```python
df4 = df3.reindex(columns=['monster_id', 'caution_level', 'species', 'habitat', 'height', 'weight'])
df4
```
