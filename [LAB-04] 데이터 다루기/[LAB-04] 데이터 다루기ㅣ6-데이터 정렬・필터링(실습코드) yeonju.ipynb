## #03. 데이터 필터링 

### 1. 숫자형 변수의 조건 검색

```python
# r1 = origin[origin['국어'] > 80]
r1 = origin.query('국어 > 80' )
r1
```

### 2. 문자열 변수의 조건 검색 

```python
# r2 = origin[origin["성별"] == '남자']
r2 = origin.query("성별 == '남자'")
r2
```

### 3. AND 연산을 사용하는 조건 검색 

```python
# r3 = origin[(origin['국어'] > 80 ) & (origin['수학'] > 80) ]
r3 = origin.query('국어 > 80 and 수학 > 80')
r3
```

### 4. OR 연산을 사용하는 조건 검색

```python
# r4 = origin[(origin['국어'] <= 80) | (origin['수학'] <= 80)]  
r4 = origin.query('국어 <= 80 or 수학 <= 80')
r4
```

### 5. 비교식에 변수를 사용하는 조건검색 
- @변수이름 형태로 조건식에 포함한다.

```python
point = int(input('점수를 입력하세요: '))
print("국어 점수가 %d점 보다 높은 학생 " % point)

r5 = origin.query('국어 > @point')
r5
```

### 6. 열 단위 필터링 

```python
fl = origin.filter(['국어', '영어'])
fl
```
