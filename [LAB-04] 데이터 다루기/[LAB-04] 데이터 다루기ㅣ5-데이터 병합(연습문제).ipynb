## 연습문제 1 - 신입 데이터 분석가를 위한 실전 미션!

### 고객 데이터 (직접 생성)

```python
customer_data = {
    'customer_id': [101, 102, 103, 104, 105, 106],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'age': [25, 32, 45, 38, 29, 41]
}
customers_df = DataFrame(customer_data)
customers_df
```

### 구매 데이터 (데이터 가져오기)

```python
purchases_df = load_data('order_history')
purchases_df
```

### 1. 구매 데이터의 price 컬럼에 있는 결측치는 전체 상품의 평균 가격으로 채워주세요.

```python
# 원본 복사
결측치처리 = purchases_df.copy()

# 전체 상품의 평균 가격
average_price = 결측치처리['price'].mean()

# 평균 가격으로 결측치 대체
# --> 특정 컬럼에 대한 처리이므로 결과는 Series 형태로 반환
결측치처리_결과 = 결측치처리['price'].fillna(average_price)

# Series 객체를 원본 데이터프레임의 price 컬럼에 대입
결측치처리['price'] = 결측치처리_결과
결측치처리
```

### 2. 정제된 두 데이터프레임을 customer_id 기준으로 병합하세요.

```python
병합 = merge(결측치처리, customers_df, on='customer_id')
병합
```

### 3. 기술통계량 도출

- customer_id를 제외하고 기술통계량을 도출해야 한다.

```python
기술통계_df = 병합.drop(columns=['customer_id', 'item', 'name'])
기술통계_df.describe().T
```

## 연습문제 2 - 흩어진 판매 데이터 통합하기

### 데이터 가져오기

```python
강남점 = load_data('shop_강남')
강남점
```

```python
종로점 = load_data('shop_종로')
종로점
```

### 1. 강남점 데이터에는 'Mouse' 상품의 가격(Price)이 누락되어 있습니다. 이 값을 8000으로 채워주세요.

```python
강남점cp = 강남점.copy()
강남점cp.loc[1, 'Price'] = 8000
강남점cp
```

### 2. 종로점 데이터에는 'Keyboard' 상품의 주문 수량(Quantity)이 999로 잘못 입력되어 있습니다. 실제 수량은 9이므로, 이 값을 수정해주세요.

```python
종로점cp = 종로점.copy()
종로점cp.loc[1, 'Quantity'] = 9
종로점cp
```

### 3, 4. 두 개의 DataFrame을 하나의 DataFrame으로 합치세요. 합치는 과정에서 기존의 인덱스는 의미가 없으므로, 새로운 인덱스를 부여하세요. 화면에 최종 데이터 프레임을 출력하세요.

```python
df_total = concat([강남점cp, 종로점cp], ignore_index=True)
df_total
```
