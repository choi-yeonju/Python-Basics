## 준비작업

### 라이브러리 참조

```python
from jussam import load_data
```

### 데이터 가져오기

```python
df = load_data("online_shop_sales")
df
```

## 요구사항 해결

### 1. 고객별 총 주문 금액 구하기

```python
df1 = df.copy()
df1["Total"] = df1["Price"] * df1["Quantity"]
df2 = df1.filter(["UserID", "Total"]).groupby("UserID").sum()
df2
```

### 2. 총 주문 금액의 내림차순으로 상위 2명

```python
df2.sort_values("Total", ascending=False).head(2)
```
