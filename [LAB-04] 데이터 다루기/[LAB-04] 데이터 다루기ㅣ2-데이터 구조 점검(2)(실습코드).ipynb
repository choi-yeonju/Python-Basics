## #02. 행,열 재배치

### 1. 열 순서 변경

- 지정된 순서대로 열이 재정렬 된 결과가 반환된다.(`inplace` 파라미터 지원 안함)
- 원본은 변화 없음. 결과가 적용된 복사본이 생성된다.
- 재배치 되지 못한 열은 제외된다.

```python
열순서_변경df = origin.reindex(columns=['국어','수학','과학','영어'])
열순서_변경df
```

### 2. 행 순서 변경

- 지정된 순서대로 행이 재정렬 된 결과가 반환된다.
- 원본은 변화 없음. 결과가 적용된 복사본이 생성된다.
- 재배치 되지 못한 행은 제외된다.

```python
행_순서변경df = origin.reindex(index=['석영','호영','민철','수현'])
행_순서변경df
```

## #03. 행 추가하기

### 1. 리스트로 행 추가하기

- 추가될 행의 인덱스 이름을 지정해야 한다.
- 리스트로 추가할 경우 DataFrame의 컬럼 순서에 맞게 지정해야 한다.
- 누락되는 값이 있거나 값의 수가 초과할 경우 에러

```python
리스트_행추가df = origin.copy()
리스트_행추가df.loc['정호'] = [2, '남자', 90, 80, 70, 60]
리스트_행추가df
```

### 2. 딕셔너리로 추가하기

-   컬럼 순서는 상관 없다.
-   누락되는 값이 있거나 값의 수가 초과할 경우 에러

```python
딕셔너리_행추가df = origin.copy()
딕셔너리_행추가df.loc['민정'] = {'성별': '여자', '학년': 4, '국어': 81,
                              '영어': 72, '수학': 84, '과학': 90}
딕셔너리_행추가df
```

### 3. Series 객체로 행 추가하기

- 순서는 상관 없다.
- Series 객체의 index가 DataFrame의 변수명과 동일해야 한다.

```python
시리즈_행추가df = origin.copy()
s = Series([2, '남자', 90, 80, 70, 60],
           index=['학년', '성별', '국어', '영어', '수학', '과학'])
시리즈_행추가df.loc['형석'] = s
시리즈_행추가df
```

### 4. 빈 값을 포함하는 행 추가하기

- 데이터를 추가할 때 값이 없는 필드는 생략할 수 있다.
- Series 형태의 객체로 추가할 때만 가능함

```python
빈값_행추가df = origin.copy()
s = Series([2, '남자', 93, 71],
           index=['학년', '성별', '국어', '수학'])
빈값_행추가df.loc['석호'] = s
빈값_행추가df
```

### 5. 기존의 행을 복사해서 행 추가 하기

```python
행_복사df = origin.copy()
행_복사df.loc['철민'] = 행_복사df.loc['철수']
행_복사df
```

### 6. 단일 값으로 행 추가하기

- 모든 변수에 같은 값이 저장된다.

```python
단일값_행추가df = origin.copy()
단일값_행추가df.loc['희영'] = 100
단일값_행추가df
```

## #04. 열 추가하기

### 1. 리스트로 열 추가하기

- 모든 행에 대한 값을 순서대로 나열한다.
- 기존의 행과 비교하여 추가되는 값이 부족하거나 초과하면 에러 발생
- 빈값을 추가하고자 하는 경우 해당 위치에 `None`이라고 명시

```python
리스트_열추가df = origin.copy()
리스트_열추가df['한국사'] = [72, 56, 90, None, 92, 90, 75, 80, 60]
리스트_열추가df
```

### 2. 딕셔너리로 열 추가하기

- 추가되지 않을 행은 명시하지 않는다.
- 데이터프레임에 존재하지 않는 인덱스가 포함될 경우 해당 값을 버린다.
- 명시되지 않은 행은 빈값(`NaN`)으로 설정된다.

```python
딕셔너리_열추가df = origin.copy()
딕셔너리_열추가df['세계사'] = {"철수": 90, "영희": 88, "민철": 72, "수현": 56,
                            "호민": 65, "용식": 77, "나영": 95, "석영": 54}
딕셔너리_열추가df
```

### 3. Series 객체로 추가하기

- 각 값이 연결될 행의 이름(index)를 지정해야 한다.
- 부분적으로 값을 비워 둘 수 있다.

```python
시리즈_열추가df = origin.copy()
s = Series([72, 56, 90, 88], index=['철수', '민철', '호영', '용식'])
시리즈_열추가df['윤리'] = s
시리즈_열추가df
```

### 4. 기존의 열을 복사해서 열 추가하기

```python
열_복사df = origin.copy()
열_복사df['체육'] = 열_복사df['과학']
열_복사df
```

### 5. 단일 값으로 열 추가하기

```python
단일_값추가df = origin.copy()
단일_값추가df['체육'] = 100
단일_값추가df
```

## #05. 행 삭제하기

### 1. 단일 행 삭제

- 추가는 원본 자체가 변하지만 삭제는 결과가 반영된 새로운 복사본이 생성된다.
- `drop()` 메서드도 `inplace=True` 파라미터 적용 가능


```python
행_삭제df = origin.copy()
행_삭제df = 행_삭제df.drop(index='철수')
행_삭제df
```

### 2. 다중 행 삭제

- 삭제할 행의 인덱스를 리스트로 지정한다.

```python
다중삭제df = origin.copy()
다중삭제결과df = 다중삭제df.drop(index=['철수','민철','수현'])
다중삭제결과df
```

### 3. index에 대한 인덱싱으로 행 삭제

```python
인덱싱_행_삭제df = origin.copy()
인덱싱_행_삭제df.index
```

```python
인덱싱_행_삭제결과df = 인덱싱_행_삭제df.drop(index=인덱싱_행_삭제df.index[0])
인덱싱_행_삭제결과df
```

### 4. index에 대한 슬라이싱으로 행 삭제

```python
다중_행_삭제df = origin.copy()
print(다중_행_삭제df.index)

다중_행_삭제결과df = 다중_행_삭제df.drop(index=다중_행_삭제df.index[0:7])
다중_행_삭제결과df
```

## #06 열 삭제

### 1. 단일 열 삭제

- 행 삭제는 `drop()` 메서드에 `index` 파라미터를 설정하고,     
  열 삭제는 `columns` 파라미터를 설정한다.

```python
단일_열_삭제df = origin.copy()
단일_열_삭제결과df = 단일_열_삭제df.drop(columns='학년')
단일_열_삭제결과df
```

### 2. 다중 열 삭제

- `columns` 파라미터에 삭제 대상 변수 이름을 리스트 타입으로 설정한다.

```python
다중_열_삭제df = origin.copy()
다중_열_삭제결과df = 다중_열_삭제df.drop(columns=['학년','성별','과학'])
다중_열_삭제결과df
```

### 3. 컬럼명에 대한 인덱싱으로 열 삭제

```python
인덱싱_열_삭제df = origin.copy()
print(인덱싱_열_삭제df.columns) # 컬럼 이름에 대한 연속성 객체

인덱싱_열_삭제결과df = 인덱싱_열_삭제df.drop(columns=인덱싱_열_삭제df.columns[0])
인덱싱_열_삭제결과df
```

### 4. 컬럼명에 대한 슬라이싱으로 열 삭제

```python
다중_열_삭제df = origin.copy()
print(다중_열_삭제df.columns)

다중_열_삭제결과df = 다중_열_삭제df.drop(columns=다중_열_삭제df.columns[1:-2])
다중_열_삭제결과df
```
