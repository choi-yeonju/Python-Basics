## #02. 문자열 데이터 처리

### 1. 대소문자 변경

- `Series.str.lower()` : 모든 문자를 소문자로 변환
- `Series.str.upper()` : 모든 문자를 대문자로 변환
- `Series.str.title()` : 각 단어의 첫 글자를 대문자로 변환
- `Series.str.capitalize()` : 문자열의 첫 글자만 대문자로 변환

```python
# 원본 데이터 프레임 복사
df = origin.copy()

# 예: 이메일 컬럼의 대소문자 조작
df['이메일_소문자'] = df['이메일'].str.lower()  # 소문자 변경
df['이메일_대문자'] = df['이메일'].str.upper()  # 대문자 변경
df['단어_첫글자'] = df['아이디'].str.title()    # 단어 첫 글자 대문자
df['첫글자'] = df['아이디'].str.capitalize()    # 문장 첫 글자 대문자
df[['이메일','이메일_소문자','이메일_대문자','단어_첫글자','첫글자']].head()
```

### 2. 공백 제거와 패딩

- `str.strip()` : 양쪽 공백 제거
- `str.lstrip()` : 왼쪽 공백 제거
- `str.rstrip()` : 오른쪽 공백 제거
- `str.zfill(width)` : 왼쪽을 0으로 채워 지정 길이로 만듭니다.
- `str.pad(width, side='left'/'right'/'both', fillchar=' ')` : 지정 길이로 패딩

```python
df['strip'] = df['메모'].str.strip()                    # 앞뒤 공백 제거
df['lstrip'] = df['메모'].str.lstrip()                  # 왼쪽 공백 제거
df['rstrip'] = df['메모'].str.rstrip()                  # 오른쪽 공백 제거

# 왼쪽 0으로 채워 지정길이로 만듦
df['zfill'] = df['아이디'].str.zfill(12)

# 지정된 길이로 만들기 위해 side에 지정한 위치(left,right,both)에 fillchar를 채움
df['pad'] = df['아이디'].str.pad(12, side='both', fillchar='*')

df[['strip','lstrip','rstrip','zfill','pad']].head()
```

### 3. 특정 패턴 포함/시작/끝 검사

- `str.contains('패턴')` : 패턴이 포함되어 있는지 검사
- `str.startswith('문자열')` : 해당 문자열로 시작하는지 검사
- `str.endswith('문자열')` : 해당 문자열로 끝나는지 검사

```python
# 이메일이 'naver.com'을 포함하는지 확인
df['네이버이메일'] = df['이메일'].str.contains('naver.com')
# 전화번호가 '010-'으로 시작하는지
df['휴대폰번호'] = df['전화번호'].str.startswith('010-')
# 주소가 '구'로 끝나는지
df['구단위주소'] = df['주소'].str.endswith('구')
# 결과확인
df[['이메일','네이버이메일','전화번호','휴대폰번호','주소','구단위주소']].head()
```

### 4. 찾기/개수/치환

- `str.find('문자')` : 처음 나오는 위치(없으면 -1)
- `str.count('문자')` : 패턴이 몇 번 나오는지
- `str.replace('패턴', '대체')` : `패턴`을 `대체`로 치환 
- 정규식을 이용해 복잡한 패턴을 대체할 수 있습니다.

```python
# 메모에서 '중' 문자가 처음 나오는 위치 (0부터 시작, 못찾으면 -1)
df['메모_중_위치'] = df['메모'].str.find('중')

# 메모에서 '중'이라는 글자의 개수
df['메모_중_개수'] = df['메모'].str.count('중')

# '메모 없음'이라는 문자열을 'empty'로 치환
df['메모없음'] = df['메모'].str.replace('메모 없음', 'empty')

df[['메모','메모_중_위치','메모_중_개수','메모없음']].head()
```

### 5. 분할/추출/결합

- `str.split(sep)` : 문자열을 분할하여 리스트를 만듭니다.
- `str.get(i)` : split 결과(리스트)에서 i번째 요소를 가져옵니다.
- `str.cat(sep=...)` : 여러 문자열 컬럼을 합칩니다.

```python
# `@`기준으로 분할한 후 0번째 요소 반환
df['사용자명'] = df['이메일'].str.split('@').str.get(0)
# `@`기준으로 분할한 후 1번째 요소 반환
df['도메인'] = df['이메일'].str.split('@').str.get(1)
# 이름과 이메일 뒷부분 결합
df['이름_이메일'] = df.index.str.cat(df['도메인'], sep='@')
# 결과확인
df[['사용자명','도메인','이름_이메일']].head()
```

### 6. 타입 검사 (숫자/문자 여부)

- `str.isdigit()` : 문자열이 숫자로만 구성되어 있으면 True (음수나 소수는 False)
- `str.isnumeric()` : 숫자 형태인지
- `str.isalpha()` : 알파벳 문자만인지
- `str.isalnum()` : 알파벳+숫자만인지

```python
df['아이디_숫자여부(1)'] = df['아이디'].str.isdigit()
df['아이디_숫자여부(2)'] = df['아이디'].str.isnumeric()
df['등급_문자여부'] = df['등급'].str.isalpha()
df['등급_문자+숫자여부'] = df['등급'].str.isalnum()

df[['아이디','아이디_숫자여부(1)', '아이디_숫자여부(2)', '등급','등급_문자여부', '등급_문자+숫자여부']].head()
```

### 7. 길이/슬라이스

```python
df['이름_길이'] = df.index.str.len()        # 문자열 길이
df['이름_성'] = df.index.str.slice(0,1) # 0번째 부터 1번째 전까지 자름
df['전화번호_뒷4자리'] = df['전화번호'].str.slice(-4)  # 마지막 4자리
df[['이름_길이','이름_성','전화번호','전화번호_뒷4자리']].head()
```

## #04. 시계열 데이터 처리

### 1. 문자열을 날짜 타입으로 변환

```python
df = origin.copy()  # 원본 데이터 보호를 위해 복사본 생성

# 문자열에 대한 날짜 타입 변환
# --> 년,월,일에 대한 format 지정 필요
df['주문일_타입변환'] = to_datetime(df['주문일'], format='%Y년 %m월 %d일')

# 변환 전,후의 데이터 타입 확인
print(df[['주문일', '주문일_타입변환']].info())

# 변환 전,후의 데이터 확인
df[['주문일', '주문일_타입변환']].head()
```

### 2. 년/월/일 정보 추출

날짜 타입의 컬럼은 `.dt` 속성을 사용하여 다양한 정보를 추출할 수 있다.

- `.dt.year` : 년도 추출
- `.dt.month` : 월 추출
- `.dt.day` : 일 추출

```python
# 주문일에서 년/월/일 정보 추출
df['주문년도'] = df['주문일_타입변환'].dt.year
df['주문월'] = df['주문일_타입변환'].dt.month
df['주문일자'] = df['주문일_타입변환'].dt.day

# 결과 확인
df[['고객명', '주문일', '주문년도', '주문월', '주문일자']].head()
```

### 3. 요일 정보 추출

```python
# 주문일 기준 요일 정보 추출
df['주문요일영문'] = df['주문일_타입변환'].dt.day_name() # 영문 요일명
df['주문요일번호'] = df['주문일_타입변환'].dt.dayofweek  # 0=월요일, 6=일요일

# 한글 요일명 매핑
요일매핑 = {0: '월요일', 1: '화요일', 2: '수요일', 3: '목요일',
           4: '금요일', 5: '토요일', 6: '일요일'}
df['주문요일'] = df['주문요일번호'].map(요일매핑)

# 결과 확인
df[['고객명', '주문일', '주문요일영문', '주문요일번호', '주문요일']].head()
```

### 4. 분기와 몇 번째 날짜인지 추출

```python
df['주문분기'] = df['주문일_타입변환'].dt.quarter        # 분기 (1~4)
df['연중몇일째'] = df['주문일_타입변환'].dt.dayofyear    # 1년 중 몇 번째 날

# 결과 확인
df[['고객명', '주문일', '주문분기', '연중몇일째']].head()
```

### 5. 시/분/초 추출

```python
df['배송완료-시간'] = df['배송완료일시'].dt.hour
df['배송완료-분'] = df['배송완료일시'].dt.minute
df['배송완료-초'] = df['배송완료일시'].dt.second

# 결과 확인
df[['고객명', '배송완료-시간', '배송완료-분', '배송완료-초']].head()
```

### 6. 날짜 간 차이 계산

- 두 날짜 간 차이: `(날짜2 - 날짜1).dt.days`
- 두 날짜 타입간의 연산은 timedelta 객체를 생성한다. 이 객체에 대한 dt 속성을 통해 두 날짜간의 차이를 구할 수 있다.

```python
df['배송소요일'] = (df['배송완료일시'] - df['주문일_타입변환']).dt.days
df[['주문일', '배송완료일시', '배송소요일']].head()
```

### 7. 경과일 구하기

파이썬의 기본 내장 모듈인 datetime 객체를 사용하여 오늘 날짜를 생성하고, 오늘 날짜에서 DataFrame의 날짜 타입 컬럼을 빼면 해당 컬럼의 날짜와 오늘 날짜의 차이를 구할 수 있다.

```python
오늘 = dt.now()
df['배송완료후_경과일'] = (오늘 - df['배송완료일시']).dt.days

df[['주문일', '배송완료일시', '배송완료후_경과일']].head()
```

### 8. 기준일로부터 년/월/일 더하기,빼기 (파이썬 내장 API 사용)

- `datetime.Timedelta(days=n)`: $n$일 더하거나 빼기

```python
# 재주문 예상일 (배송완료 후 30일)
df['재주문예상일'] = df['배송완료일시'] + td(days=30)

# 결과확인
df[['고객명', '배송완료일시', '재주문예상일']].head()
```

### 9. 기준일로부터 년/월/일 더하기,빼기 (Pandas API 사용)

- `pd.DateOffset(years=y, months=m, days=d, hours=h, minutes=i, seconds=s)`

```python
# 1년 A/S 보장 종료일 (주문일 기준)
df['AS보장만료일'] = df['주문일_타입변환'] + DateOffset(years=1)
# 예상 매출 집계일 (주문일에서 3개월 후)
df['매출집계예정일'] = df['주문일_타입변환'] + DateOffset(months=3)
# 리뷰 요청 예정일시 (배송완료시점에서 7일 12시간 경과 후, 혼합사용 가능)
df['리뷰요청예정'] = df['배송완료일시'] + DateOffset(days=3, hours=12)
# 결과확인
df[['고객명','주문일','AS보장만료일','매출집계예정일','리뷰요청예정']].head()
```
