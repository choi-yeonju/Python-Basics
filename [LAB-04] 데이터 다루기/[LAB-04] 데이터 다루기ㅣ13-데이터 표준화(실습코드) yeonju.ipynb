## #01. 준비작업 

### 1. 라이브러리 참조 

```python
from jussam import load_data 

# 스케일링을 위해 사용 
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler, RobustScaler
```

### 2. 샘플 데이터 가져오기 

```python
origin = load_data('nursing_grades')
origin
```

## #02. Min-Max Scaler (정규화)

```python
minmax_df = origin.copy()

x = minmax_df['필기점수']
xmin = x.min()
xmax = x.max()

minmax_df['필기점수_MinMax(1)'] = (x-xmin)/(xmax -xmin)
minmax_df.head()
```

### 2. Sklearn 활용 
- 특정 필드만 추출 

```python
x = minmax_df[['필기점수']]
x.head()
```

### 2. Sklearn 활용
- 2) 스케일링 적용 

```python
scaler = MinMaxScaler()
minmax_df['필기점수_MinMax(2)'] = scaler.fit_transform(x) 
minmax_df.head()
```

## #03. Max - Abs Scaler 

```python
abs_df = origin.copy()
x = abs_df['필기점수']
abs_df['필기점수_MaxAbs(1)'] = x / x.abs().max()
abs_df.head()
```

### 2. Sklearn 활용 

```python
scaler = MaxAbsScaler() 
x = abs_df[['필기점수']]
abs_df['필기점수_MaxAbs(2)'] = scaler.fit_transform(x)
abs_df.head()
```

## #04. Standard Scaler (z-score, 표준화)

```python
std_df = origin.copy()
x = origin['학부성적']
std_df['학부성적_Standard(1)'] = (x - x.mean()) / x.std()
std_df.head()
```

### 2. Sklearn 활용 

```python
scaler = StandardScaler()
x = std_df[['학부성적']]
std_df['학부성적_Standard(2)'] = scaler.fit_transform(x)
std_df.head()
```

## #05. Robust Scaler 

```python
rb_df = origin.copy()

x = rb_df['병원경력']
m = x.median()
iqr = x.quantile(0.75) - x.quantile(0.25)

rb_df['병원경력_Robust(1)'] = (x-m) / iqr 
rb_df.head()
```

### 2. Sklearn 활용 

```python
scaler = RobustScaler()
x = rb_df[['병원경력']]
rb_df['병원경력_Robust(2)'] = scaler.fit_transform(x)
rb_df.head()
```

## #01. 준비작업

### 1. 라이브러리 참조

```python
from jussam import load_data

# 스케일링을 위해 사용
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler, RobustScaler
```

### 2. 데이터 가져오기

```python
performance_df = load_data("performance");
performance_df
```

## 요구사항 해결

### 요구사항에 따른 스케일링 수행

```python
df = performance_df.copy()

# MinMaxScaler
df['MonthlySales_scaled'] = MinMaxScaler().fit_transform(df[['MonthlySales']])

# MaxAbsScaler
df['CustomerSatisfaction_scaled'] = MaxAbsScaler().fit_transform(df[['CustomerSatisfaction']])

# StandardScaler
scaled_standard_features = StandardScaler().fit_transform(df[['ProjectHours', 'InnovationScore']])
df['ProjectHours_scaled'] = scaled_standard_features[:, 0]
df['InnovationScore_scaled'] = scaled_standard_features[:, 1]

df.head()
```

### 스케일링 값의 합계 구하기

```python
df['TotalScaledScore'] = df['MonthlySales_scaled'] + df['CustomerSatisfaction_scaled'] + df['ProjectHours_scaled'] + df['InnovationScore_scaled']
df.head()
```

### 합계 결과가 가장 큰 직원 찾기

```python
max_data = df.sort_values('TotalScaledScore', ascending=False).head(1)
max_data
```

### 결과

```python
print(f"스케일링된 값의 합계가 가장 높은 직원의 EmployeeID는 {max_data.index[0]}입니다.")
```
