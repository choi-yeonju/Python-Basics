## 준비작업

### 라이브러리 참조

```python
from jussam import load_data
```

### 데이터 가져오기

```python
df = load_data('vip_customer')
df
```

## 요구사항 처리

### 조건에 따른 필터링

```python
# 조건 1: 총 구매액 100,000원 이상
# 조건 2: 서울 거주 고객
df2 = df.query("TotalSpending >= 100000 and City == '서울'")
df2
```

### 보고할 내용 생성

```python
# 총 지출액 기준 내림차순 정렬
sorted_vips = df2.sort_values(by='TotalSpending', ascending=False)

# 상위 5명 고객 추출
top5_vips = sorted_vips.head(5)

# 최종 보고에 필요한 'Name', 'City', 'TotalSpending' 열 선택
final_report = top5_vips.filter(['Name', 'City', 'TotalSpending'])

# 결과 출력
final_report
```
