## #02. 파생변수 추가하기

### 1. 컬럼 연산을 통한 새로운 컬럼 추가

```python
df1 = origin.copy()
df1["직접합계"] = df1["국어"] + df1["영어"] + df1["수학"] + df1["과학"]
df1
```

### 2. 기술통계값 관련 함수를 사용한 파생변수

#### 1) 컬럼 이름에 대한 슬라이싱 접근

- `axis=1`: x축
- `axis=0`: y축

```python
df2 = origin.copy()
df2["합계"] = df2[df2.columns[2:6]].sum(axis=1)
df2
```

#### 2) 인덱스 기반 위치로 슬라이싱하여 합계변수 추가

```python
df3 = origin.copy()
df3["합계"] = df3.iloc[:, 2:].sum(axis=1)
df3
```

#### 3) 나머지 기술통계 값 추가

```python
df3["평균"] = df3.iloc[:, 2:6].mean(axis=1)
df3["최대값"] = df3.iloc[:, 2:6].max(axis=1)
df3["최소값"] = df3.iloc[:, 2:6].min(axis=1)
df3["표준편차"] = df3.iloc[:, 2:6].std(axis=1)
df3["분산"] = df3.iloc[:, 2:6].var(axis=1)
df3
```

## #03. 조건별 파생변수

### 1. `if~else`

#### 1) 실습을 위해 합계와 평균에 대한 파생변수 추가

```python
df4 = origin.copy()
df4["합계"] = df4.iloc[:, 2:6].sum(axis=1)
df4["평균"] = df4.iloc[:, 2:6].mean(axis=1)
df4
```

#### 2) NumPy의 where 함수를 사용한 조건부 파생 변수

```python
df4["결과1"] = np.where(df4["평균"] >= 80, "합격", "불합격")
df4
```

#### 3) 표현식을 사용한 평균값에 따른 조건부 파생변수

```python
df4["결과2"] = "불합격"
df4.loc[df4["평균"] >= 80, "결과2"] = "합격"
df4
```

### 2. 경우의 수에 따른 조건값 (`elif`)- numpy의 `select` 함수 사용

```python
# 학점을 부여하기 위한 점수의 구간을 설정하는 조건들을 리스트로 설정
conditions = [
    (df4["평균"] >= 90),  # if
    (df4["평균"] >= 80),  # elif
    (df4["평균"] >= 70),  # elif
]
# 조건에 따라 부여될 학점
value = ["A", "B", "C"]
#                        조건        값        else
df4["학점1"] = np.select(conditions, value, default="F")
df4
```

### 3. 사용자 정의 함수 사용

#### 1) 함수 정의하기

```python
def getScore(x):
    if x >= 90:
        score = "A"
    elif x >= 80:
        score = "B"
    elif x >= 70:
        score = "C"
    else:
        score = "F"

    return score
```

#### 2) 함수의 리턴값으로 파생변수 추가하기

```python
df4["학점2"] = df4["평균"].apply(getScore)
df4
```

#### 3) 함수 정의 없이 람다식으로 익명함수 적용하기

```python
df4["학점3"] = df4["평균"].apply(
    lambda x: {10: "A", 9: "A", 8: "B", 7: "C"}.get(x // 10, "F")
)
df4
```

### 4. `map()` 메서드 활용

- 앞의 방법처럼 연산을 사용할 수는 없고 값이 완전히 일치하는 경우에만 사용할 수 있음.

```python
df4["성별코드"] = df4["성별"].map({"남자": "M", "여자": "F"})
df4
```

### 5. Pandas의 cut 함수 사용

구간에 따른 파생변수 생성

- `0`이상 `70`미만 : **F**
- `70`이상 `80`미만 : **C**
- `80`이상 `90`미만 : **B**
- `90`이상 `100`이하 : **F** (마지막 구간은 **이하**)

```python
df4["학점4"] = cut(
    df4["평균"],  # 대상 컬럼
    bins=[0, 70, 80, 90, 100],  # 구간 설정
    labels=["F", "C", "B", "A"],  # 구간별 값 설정
)
df4
```
