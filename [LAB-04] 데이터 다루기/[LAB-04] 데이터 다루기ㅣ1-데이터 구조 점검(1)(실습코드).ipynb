## #03. 데이터 타입 변경

### 1. 작업을 위한 복사본 생성

```python
컬럼타입변경df = origin.copy()
컬럼타입변경df.info()
```

### 2. 컬럼 단위 변경

- **컬럼 단위의 데이터 타입 변경**
  - 하나의 컬럼은 시리즈 개체이므로 `astype()` 함수로 타입 변경 가능.
  - 변경된 결과는 다시 원래의 컬럼에 저장해야 DataFrame에 반영된다.

```python
컬럼타입변경df['국어'] = 컬럼타입변경df['국어'].astype('float')
컬럼타입변경df.info()
```

### 3. 타입 일괄 변경 후 새로운 DataFrame 리턴 받기

- **타입 변환 적용**
  - 타입 변환을 수행할 변수만 `이름:타입` 형식의 딕셔너리로 구성한다.
  - 결과가 적용된 복사본이 리턴된다.

```python
타입변환df = origin.astype({'국어': 'float', '영어': 'float',
                           '수학': 'str', '과학': 'bool'})
타입변환df.info()
```

### 4. 타입 변환 전 원본 확인

- DataFrame 객체에 대한 `astype()`함수는 결과가 적용된 복사본을 리턴한다.     
  (원본은 변하지 않음)


```python
origin.info()
```

### 5. 범주형 데이터를 명목척도로 변환 (1)

- 컬럼 단위 변환

```python
컬럼단위_명목df = origin.copy()
컬럼단위_명목df['학년'] = 컬럼단위_명목df['학년'].astype('category')
컬럼단위_명목df['성별'] = 컬럼단위_명목df['성별'].astype('category')
컬럼단위_명목df.info()
```

### 6. 범주형 데이터를 명목척도로 변환 (2)

- DataFrame 전체에 적용 후 복사본 리턴 받기

```python
범주형df = origin.astype({'학년': 'category', '성별': 'category'})
범주형df.info()
```

### 7. 범주형 데이터를 서열척도로 변환 (1)

- **컬럼 단위로 적용하여 원본에 직접 적용하기**
  - `CategoricalDtype` 클래스의 객체를 생성하여 범주의 특성을 설정한다.

```python
컬럼단위_서열척도df = origin.copy()
컬럼단위_서열척도df['학년'] = 컬럼단위_서열척도df['학년'].astype(
        CategoricalDtype(categories=[1, 2, 3, 4], ordered=True))
컬럼단위_서열척도df.info()
```

### 8. 범주형 데이터를 서열척도로 변환 (2)

- DataFrame 전체에 적용 후 복사본 반환 받기

```python
서열척도df = origin.astype({
    "학년": CategoricalDtype(categories=[1,2,3,4], ordered=True)})
서열척도df.info()
```

## #04. 행, 열의 이름 변경

### 1. `rename()` 메서드의 사용

-  `기존이름: 새이름` 형식의 딕셔너리로 지정
-  `columns`와 `index`중 변경을 원하는 파라미터 하나만 설정 가능.
-  원본은 변화 없음. 결과가 적용된 복사본이 생성된다.
-  이름이 변경되지 않은 항목이 결과에서 제외되지는 않는다.
-  원본에 즉시 반영해야 할 경우 `inplace=True` 파라미터를 적용

```python
이름변경df = origin.rename(
    columns={'국어': 'kor', '영어': 'eng', '수학': 'math', '과학': 'sci'},
    index={'석영': 'suk', '호영': 'ho', '민철': 'min', '수현': 'su'}
)
이름변경df
```
