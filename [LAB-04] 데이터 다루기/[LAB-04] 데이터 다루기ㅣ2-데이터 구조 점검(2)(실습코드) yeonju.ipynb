## #03. 행 추가하기 

### 1. 리스트로 행 추가하기 

```python
리스트_행추가df = origin.copy()
리스트_행추가df.loc['정호'] = [2, '남자', 90, 80, 70, 60]
리스트_행추가df
```

### 2. 딕셔너리로 행 추가하기

```python
딕셔너리_행추가df = origin.copy()
딕셔너리_행추가df.loc['민정'] = {'성별':'여자', '학년':4, '국어':81, '영어':72, '수학':84, '과학':90}
딕셔너리_행추가df
```

### 3. Series 객체로 행 추가하기 

```python
시리즈_행추가df = origin.copy()
s = Series([2, '남자', 90, 80, 70, 60], 
           index = ['학년', '성별', '국어', '영어', '수학', '과학'])
시리즈_행추가df.loc['형석'] = s
시리즈_행추가df
```

### 4. 빈 값을 포함하는 행 추가하기 

```python
빈값_행추가df = origin.copy()
s = Series( [2, '남자', 93, 71], 
           index = ['학년', '성별', '국어', '수학'])
빈값_행추가df.loc['석호'] = s 
빈값_행추가df
```

### 5. 기존의 행을 복사해서 행 추가하기 

```python
행_복사df = origin.copy()
행_복사df.loc['철민'] = 행_복사df.loc['철수']
행_복사df
```

### 6. 단일 값으로 행 추가하기 

```python
단일값_행추가df = origin.copy()
단일값_행추가df.loc['희영'] = 100 
단일값_행추가df
```
