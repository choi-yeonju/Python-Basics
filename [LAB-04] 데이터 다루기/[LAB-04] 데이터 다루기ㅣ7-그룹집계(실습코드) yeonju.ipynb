## #04. 두 개 이상의 변수에 대한 그룹화

### 1. 그룹화 할 기준이 두 개 이상인 경우 

```python
origin.groupby(['년도', '월']).mean()
```

### 2. 두 개 이상의 집계함수를 사용할 경우 

```python
origin.filter(['년도', '발생건수']).groupby('년도').agg(
    ['sum', 'max', 'min', 'mean']
)
```

### 3. 컬럼별로 집계함수를 다르게 지정하는 경우 

```python
origin.drop("월", axis=1).groupby('년도').agg(
    {"발생건수":['mean', 'sum'], '사망자수':'max', '부상자수':'min'}
)
```

## #05. 사용자 정의 함수를 사용한 집계

```python
def myFunction(x):
    # x는 Series객체 - 컬럼 하나가 통째로 전달됨 
    return x.max() - x.min()
```

```python
# 사용자 정의 함수 이름은 문자열이 아님 
origin.drop('월', axis=1).groupby('년도').agg(['max', 'min', myFunction])
```

### 2. 람다식 적용 

```python
origin.drop('월', axis=1).groupby('년도').agg(
    ['max', 'min', lambda x : x.max() - x.min()])
```

## #06. 인덱스 지정 해제 

### 1. groupby() 메서드에서 사용한 필드는 기본적으로 결과집합의 인덱스로 지정됨

```python
origin.drop('월', axis=1).groupby('년도', as_index=False).sum()
```

## #1. 연습문제 - 온라인 쇼핑몰의 다음 분기 프로모션 전략 수립 

### 1. 결측치 확인 / price 컬럼에 결측치는 전체 상품의 평균 price로 대체

```python
df = load_data('shop_order_list')
df
```

```python
df.isnull().sum()
```

```python
df2 = df.copy()
price_mean = df2['Price'].mean()
df2['Price'] = df2['Price'].fillna(price_mean)
print(df2.isnull().sum())
```

```python
top_categories = df2['Category'].value_counts().head(3)
print("가장 많은 주문이 발생한 카테고리 TOP 3: ")
top_categories
```

```python
region_avg_price = df2.filter(['Region', 'Price']).groupby('Region').mean().sort_values('Price', ascending=False)
region_avg_price
```
