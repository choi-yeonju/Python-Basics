## #03. 결측치 소거

### 1. 행 단위 삭제

결측치가 있는 모든 행 삭제 (원본은 변화 없음, 삭제 결과가 리턴됨)

`inplace=True` 파라미터를 적용할 경우 원본에 즉시 반영되고 리턴값 없음

```python
na1 = origin.dropna()
na1
```

### 2. 열 단위 삭제

`dropna()` 메서드에 `axis=1` 파라미터를 적용한다. (`기본값=0`)

```python
na2 = origin.dropna(axis=1)
na2
```

### 3. 특정 열에 대해서만 결측치가 존재하는 행 삭제

```python
na3 = origin.dropna(subset=['kor'])
na3
```

## #04. 결측치 대체

### 1. 고정값으로 대체

모든 결측치를 동일한 값으로 교체

원본은 변화 없음, 삭제결과 리턴됨

`inplace=True`를 적용할 경우 원본에 즉시 반영되고 리턴값 없음

```python
re_df1 = origin.fillna(value=50)
re_df1
```

### 2. 통계적 값으로 대체 (1) - 결측치를 정제할 규칙을 담고 있는 객체 생성

각 열단위로 평균`(strategy='mean')`을 결측치`(missing_values)`에 지정

strategy 옵션 : mean=평균, median=중앙값, most_frequent: 최빈값(가장 많이 관측되는 수)

> 이 객체의 생성이 결측치를 즉시 처리했다는 것을 의미하는 것은 아니다.

```python
imr = SimpleImputer(missing_values=np.nan, strategy='mean')
imr
```

### 3. 통계적 값으로 대체 (2) - 생성된 규칙을 적용

결측치가 정제된 2차원 배열로 리턴되므로 다시 DataFrame으로 재구성할 필요가 있다.

```python
df_imr = imr.fit_transform(origin.values)
df_imr
```

### 4. 통계적 값으로 대체 (3) - 결측치 정제 결과를 DataFrame으로 재구성

```python
re_df2 = DataFrame(df_imr,
                   index=origin.index,      # 원본의 인덱스
                   columns=origin.columns   # 원본의 컬럼명
                )
re_df2
```

### 5. 특정 컬럼 하나만 처리하기 (1)- 컬럼의 값만 추출한 후 n행1열로 재구성

- sklearn의 SimpleInputer는 파라미터로 2차 배열을 요구한다.
- 특정 컬럼 하나에 대해서만 결측치 대체를 수행해야 할 경우    
  sklearn의 요구사항에 맞춰 n행 1열로 재구성해야 한다.

```python
reshape = origin['kor'].values.reshape(-1, 1)
reshape
```

### 6. 특정 컬럼의 값 하나만 처리하기 (2) - 결측치 처리하기

```python
reshape_imr = imr.fit_transform(reshape)
reshape_imr
```

### 7. 특정 컬럼의 값 하나만 처리하기 (3) - 결측치 처리 결과를 DataFrame의 컬럼에 적용하기

```python
특정컬럼_정제df = origin.copy()
특정컬럼_정제df['kor'] = reshape_imr
특정컬럼_정제df
```
