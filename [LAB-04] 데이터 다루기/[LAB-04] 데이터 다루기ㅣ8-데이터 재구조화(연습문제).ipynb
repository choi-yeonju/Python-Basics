## 연습문제 2 - 동네 아이스크림 가게 매출 분석 챌린지!

```python
df = load_data('icecream_sales')
df
```

### 1. 구조와 정보를 확인, 결측치가 있는지 확인, 결측치 행을 제거

#### 1) 구조와 정보 확인

```python
df.info()
```

#### 2) 결측치가 있는지 확인

```python
df.isna().sum()
```

#### 3) 결측치 행 제거

```python
df2 = df.dropna()
df2
```

#### 2. 피벗 테이블

```python
df3 = pivot_table(df2, index='Date', columns='Flavor', values='Quantity')
df3
```

### 3. 피벗 테이블을 다시 Long format으로 분리

```python
df4 = df3.reset_index()
df4
```

```python
df5 = melt(df4, id_vars=['Date'],
           value_vars=['딸기','민트초코','바닐라','초콜릿'],
           var_name='Flavor', value_name='Quantity')
df5
```

## 연습문제 3 - 게임 토너먼트 최고 득점자 찾기

### 데이터 가져오기

```python
df = load_data("game_scores")
df
```

### 1. 각 스테이지 점수가 열로 구성된 데이터를, 선수와 스테이지 레벨에 따른 점수를 나타내는 형태로 재구조화하세요.

```python
# 요구사항에 스테이지에 모두 참여한 선수만을 대상으로 하라고 했으므로 결측치 행을 제거해야 한다.
df1 = df.dropna()

df2 = df1.set_index('player')
df3 = DataFrame(df2.stack())
df3
```

### 2. 선수별 평균 점수를 계산하세요.

```python
df4 = df3.reset_index()
df4.filter(['player', 0]).groupby('player').mean()
```
