## #02. 데이터 정렬

### 1. 변수에 의한 정렬

- 원본에 즉시 반영해야 할 경우 `inplace=True` 파라미터를 적용
- `ascending=False` 파라미터는 내림차순 정렬을 설정함 (기본값=True, 오름차순)

```python
단일정렬df = origin.sort_values('학년', ascending=False)
단일정렬df
```

### 2. 여러 변수 정렬

- 정렬 기준이 되는 컬럼 이름을 리스트로 설정하면 첫 번째 컬럼의 값이 동일한 경우 두 번째 컬럼을 기준으로 정렬한다.
- 각 변수에 대한 정렬 옵션 역시 리스트로 설정한다.

```python
복수정렬df = origin.sort_values(['학년', '국어'],
                            ascending=[False, False])
복수정렬df
```

### 3. 인덱스를 기준으로 하는 정렬

- 인덱스는 하나만 존재하므로 열 이름을 지정할 필요가 없다.
- 정렬 방향을 위한 `ascending` 파라미터를 `False`로 설정하면 내림차순으로 정렬된다.
- `inplace=True`를 적용할 경우 원본에 즉시 반영된다.

```python
인덱스정렬df = origin.sort_index(ascending=False)
인덱스정렬df
```

## #03. 데이터 필터링

### 1. 숫자형 변수의 조건 검색

```python
# r1 = origin[origin['국어'] > 80]
r1 = origin.query('국어 > 80')
r1
```

### 2. 문자열 변수의 조건 검색

```python
#r2 = origin[origin["성별"] == '남자']
r2 = origin.query("성별 == '남자'")
r2
```

### 3. AND 연산을 사용하는 조건 검색

```python
#r3 = origin[(origin['국어'] > 80) & (origin['수학'] > 80)]
r3 = origin.query('국어 > 80 and 수학 > 80')
r3
```

### 4. OR 연산을 사용하는 조건 검색

```python
#r4 = origin[(origin['국어'] <= 80) | (origin['수학'] <= 80)]
r4 = origin.query('국어 <= 80 or 수학 <= 80')
r4
```

### 5. 비교식에 변수를 사용하는 조건 검색

- `@변수이름` 형태로 조건식에 포함한다.

```python
point = int(input('점수를 입력하세요: '))
print("국어 점수가 %d점 보다 높은 학생" % point)

r5 = origin.query('국어 > @point')
r5
```

### 6. 열 단위 필터링
