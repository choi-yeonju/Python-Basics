## #02. 그래프 기본 설정

#### 1. 폰트 파일 확인

- "#02"의 코드는 ipynb 파일 하나당 한번만 수행하면 된다.
  - 같은 내용의 코드가 다시 실행되기 전까지 현재 소스파일의 모든 그래프에 전역으로 적용되는 설정.

```python
# 한글을 지원하는 폰트 파일의 경로
fpath = "./helpers/fonts"

# 폰트 파일의 경로를 리스트로 가져오기
font_files = gl.glob(os.path.join(fpath, "*.ttf"))
font_files
```

### 2. 폰트 파일 적용

```python
for f in font_files:
    fm.fontManager.addfont(f)             # 폰트 등록
    fprop = fm.FontProperties(fname=f)    # 폰트의 속성을 읽어옴
    fname = fprop.get_name()              # 읽어온 속성에서 폰트의 이름만 추출
    plt.rcParams['font.family'] = fname   # 그래프에 한글 폰트 적용
```

### 3. 그래프 기본 옵션

```python
my_dpi = 200                                # 이미지 선명도(100~300)
plt.rcParams['font.size']   = 12            # 기본 폰트 크기
plt.rcParams['axes.unicode_minus'] = False  # 그래프에 마이너스 깨짐 방지
plt.rcParams['figure.dpi'] = my_dpi 		# 그래프의 dpi 설정
plt.rcParams['savefig.dpi'] = my_dpi 		# 저장되는 그래프의 dpi 설정
plt.rcParams['lines.linewidth'] = 2 		# 그래프 선 굵기 설정
plt.rcParams['axes.axisbelow'] = True       # 그래프의 축과 격자선을 선보다 뒤에 배치
```

## #03. 그래프 기본 코드 구성

### 1. 모든 그래프에 공통적으로 적용되는 기본 템플릿

- 그래프의 크기는 inch 단위로 지정되어야 한다.
    - $inch = px / 100$

```python
# 1) 그래프 초기화
width  = 1280                       # 그래프 가로 px 크기
height = 640                        # 그래프 세로 px 크기
rows = 1                            # 그래프 행 수
cols = 1                            # 그래프 열 수
my_figsize = (width / 100, height / 100)
fig, ax = plt.subplots(rows, cols, figsize=my_figsize, dpi=my_dpi)
ax.grid(True, alpha=0.5)           # 배경 격자 표시 (alpha는 투명도)

# 2) 그래프 그리기
# ... 이 부분에 그래프를 그리는 코드가 들어갑니다. 예시에서는 생략

# color: 글자 색상 (기본값은 black)
# fontsize: 글자 크기 (기본값은 plt.rcParams['font.size'] 설정을 따름)
# fontweight: 글자 굵기 (100~1000 사이 100단위 값) / pad, labelpad : 그래프와 텍스트의 간격
ax.set_title("제목", color="#0066ff", fontsize=24, fontweight=700, pad=15)
ax.set_xlabel("X축", color="#ff6600", fontsize=16, fontweight=500, labelpad=5)
ax.set_ylabel("Y축", color="#ff00ff", fontsize=16, fontweight=500, labelpad=5)

# 4) 출력
plt.tight_layout()                       # 외부 여백 제거
plt.savefig("myplot.png", dpi=my_dpi)    # 이미지 저장(생략 가능)
plt.show()                               # 그래프 화면 출력
plt.close()                              # 그래프 작업 종료
```
