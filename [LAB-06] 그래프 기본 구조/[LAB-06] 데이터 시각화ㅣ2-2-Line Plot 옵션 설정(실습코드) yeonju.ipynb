## #02. Seaborn 패턴별 LinePlot 

### 1. 자동 인식 패턴 

```python
# 그래프 초기화 
fig, ax = my_plot.init(title="직선 그리기", xlabel="X", ylabel="Y")

# 그래프 그리기 (직선 그리기) 
sb.lineplot([10, 20, 30, 40, 50])

# 출력 
my_plot.show()
```

### 2. 개별 데이터 직접 지정 패턴 

```python
# 그래프 초기화 
fig, ax = my_plot.init(title="개별 데이터 지정", xlabel="X", ylabel="Y")

# 그래프 그리기 
sb.lineplot(x=[2,3,4,5,6], y=[10,20,30,40,50])

# 출력 
my_plot.show()
```

### 3. DataFrame + 컬럼명 패턴을 위한 데이터 불러오기 

```python
origin = load_data('icecream_sales')
origin
```

### 4. DataFrame + 컬럼명 패턴 

```python
# 그래프 초기화 
fig, ax = my_plot.init(title="아이스크림 판매량", xlabel="날짜", ylabel="판매수량")


# LinePlot 그리기 
sb.lineplot(data=origin, x='Date', y="Quantity")

# 출력 
my_plot.show()
```

### 5. DataFrame + 컬럼명 패턴 + hue 파라미터 적요 

```python
# 그래프 초기화 
fig, ax = my_plot.init(title="종류별 아이스크림 판매량", xlabel="날짜", ylabel="판매수량")

# LinePlot 그리기 
sb.lineplot(data = origin, x='Date', y='Quantity', hue="Flavor")

# 출력 
my_plot.show()
```

### 6. 색상 팔레트 적용 

```python
# 그래프 초기화 
fig, ax = my_plot.init(title="종류별 아이스크림 판매량", 
                       xlabel="날짜", ylabel="판매수량")
# LinePlot 그리기 
sb.lineplot(data=origin, x='Date', y="Quantity", 
            hue="Flavor", palette="PuRd")

# 출력 
my_plot.show()
```
