## #02. 2차원 배열

### 1. 2차원 리스트의 ndarray 변환

- 배열의 정보를 확인할 수 있는 프로퍼티가 포함된다.

```python
myarray = np.array([
    [98, 72, 80, 64],
    [88, 90, 80, 72],
    [92, 88, 82, 76]
])

print(myarray)

print("차원크기:", myarray.ndim)
print("차원별 원소수:", myarray.shape)
print("원소의 데이터 타입:", myarray.dtype)
```

### 2. boolean 타입으로 인덱싱

- 조건에 맞는 원소를 1차원으로 반환한다.

```python
bool_array = np.array([
    [True, False, True, False],
    [True, True, True, False],
    [True, True, True, False]
])

# 조건에 맞는 항목만 1차 배열로 추출
# myarray는 이전 실습에서 생성한 2차원 배열
result1 = myarray[bool_array]
result1
```

## #03. 비교식에 의한 검색

### 1. 조건에 맞는 원소를 1차원 배열로 반환한다.

```python
result2 = myarray[myarray >= 80]
result2
```

### 2. `AND` 검색

```python
result3 = myarray[np.logical_and(myarray >= 80, myarray <= 90)]
result3
```

### 3. `OR` 검색

```python
result4 = myarray[np.logical_or(myarray < 80, myarray > 90)]
result4
```

## #04. 정형화된 배열 생성

- 대각선 원소가 1이고 그 외에는 0인 배열

## #07. 랜덤 시드

### 1. 랜덤 시드(𝑅𝑎𝑛𝑑𝑜𝑚 𝑆𝑒𝑒𝑑)란?

- 난수(𝑟𝑎𝑛𝑑𝑜𝑚 𝑛𝑢𝑚𝑏𝑒𝑟) 는 겉보기엔 무작위지만,    
  실제로는 컴퓨터가 특정 규칙(알고리즘) 에 따라 만들어내는 가짜 무작위값이다.
  - 미리 준비된 수 많은 숫자 묶음 중 하나를 선택하고,    
    선택된 숫자 묶음의 숫자를 순차적으로 꺼내보는 것으로 이해하자.
- 이때 랜덤 시드(Random Seed) 는 그 규칙을 어디서부터 시작할지 정하는 ‘출발점 값’ 이다.
  - 같은 시드를 주면 👉 항상 같은 난수 결과가 나온다. (항상 같은 숫자 묶음을 사용)
  - 다른 시드를 주면 👉 완전히 다른 난수 결과가 나온다. (매번 다른 숫자 묶음을 사용)


### 2. 랜덤 시드를 고정하고 정규 분포를 따르는 랜덤값 생성하기

- `평균 = 0`, `표준편차 = 1`, `배열크기 = 3행 4열`

```python
np.random.seed(1234)
np.random.normal(0, 1, (4, 3))
```
