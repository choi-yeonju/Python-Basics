## #02. 2차원 배열 
### 1. 2차원 리스트의 ndarray 변환
#### - 배열의 정보를 확인할 수 있는 프로퍼티가 포함된다

```python
myarray = np.array([
    [98, 72, 80, 64],
    [88, 90, 80, 72], 
    [92, 88, 82, 76]
])

print(myarray)

print("차원크기:", myarray.ndim)
print("차원별 원소수", myarray.shape)
print("원소의 데이터 타입:", myarray.dtype)
```

### 2. boolean 타입으로 인덱싱 
#### - 조건에 맞는 원소를 1차원 배열로 반환한다.

```python
bool_array = np.array([
    [True, False, True, False],
    [True, True, True, False],
    [True, True, True, False]
])

# 조건에 맞는 항목만 1차 배열로 추출
# myarray는 이전 실습에서 생성한 2차원 배열 
result1 = myarray[bool_array]
result1
```

## #03. 비교식에 의한 검색 

```python
# 1. 조건에 맞는 원소를 1차원 배열로 반환한다. 
result2 = myarray[myarray >=80]
result2
```

## # 07. 랜덤 시드
#### 1. 랜덤 시드(Random Seed)란?
* 난수(random number)는 겉보기엔 무작위지만, 실제로는 컴퓨터가 특정규칙(알고리즘)에 따라 만들어내는 가짜 무작위 값이다. 
    * 미리 준비된 수 많은 숫자 묶음 중 하나를 선택하고, 선택된 숫자 묶음의 숫자를 순차적으로 꺼내보는 것으로 이해하자
#### 2. 랜덤 시드를 고정하고 정규 분포를 따르는 랜덤값 생성하기
* 이때 랜덤 시드(Random Seed)는 그 규칙을 어디서부터 시작할지 정하는 '출발점 값'이다.
    * 같은 시드를 주면, 항상 같은 난수 결과가 나온다. (항상 같은 숫자 묶음을 사용)
    * 다른 시드를 주면, 완전히 다른 난수 결과가 나온다.(매번 다른 숫자 묶음을 사용)

```python
# 랜덤 시드를 고정하고 정규분포를 따르는 랜덤값 생성하기
np.random.seed(1234)
np.random.normal(0,1,(4,3))
```

## #1. 연습문제 - 성적 처리 시스템 만들기 

정수형 ndarray로 변환하시오
점수가 80점 이상인 학생의 점수만 추출하여 출력하시오

모든 학생에게 다음 기준으로 학점을 부여하고 결과를 출력하시오.

90점이상 A / 80점 이상 B / 70점 이상 C / 그 외 F 

```python
import numpy as np
score = np.array([72.5, 88.0, 91.5, 67.0, 80.0, 59.5])


print("원본 점수:", score)
```

```python
result1 = score[score >= 80]

print("80점 이상:", result1)
```

```python
#  float → int 변환은 **반올림이 아니라 버림(truncate)**
```

```python
result1 = np.array(score, dtype='int')
print("원본 점수:", result1)

print("80점 이상:", result1[result1 >= 80])
```

```python
score_int = score.astype(int)
print("원본 점수:", score_int)

result = score_int[score_int >= 80]
print("80점 이상:", result)
```

```python
exam1 = [ (score>=90 ),
         (score >=80), 
         (score >=70)]
value = ['A', 'B', 'C']

result = np.select(exam1, value, default='F')
print("학점 결과", result)
```

```python
grades = np.where(score_int>=90, 'A',
          np.where(score_int>=80, 'B',
          np.where(score_int>=70, 'C', 'F')))
```
