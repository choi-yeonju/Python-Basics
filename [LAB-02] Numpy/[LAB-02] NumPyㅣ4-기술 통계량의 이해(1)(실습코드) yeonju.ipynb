## #01. 기술통계량의 이해 

### 1. 준비작업

```python
# 1. 라이브러리 참조 
import numpy as np 

# 2. 샘플 데이터 생성 
myarr = np.array([98, 72, 80, 64, 88, 65, 79, 59, 81, 74])
myarr
```

## #02.데이터의 위치, 분포를 나타내는 기술 통계량 

```python
# 1. 총 도수 
print("총 도수", len(myarr))

# 2. 최대(max), 최소(min), 합계(sum)
print("최대값:", np.max(myarr))
print("최소값:", np.min(myarr))
print("합계:", np.sum(myarr))
```

```python
# 3. 평균 
print("평균(average):", np.average(myarr))
print("평균(mean):", np.mean(myarr))
```

```python
# 4. 중앙값(직접 구하기)
size = len(myarr) # 전체 데이터 수 
sorted_arr = np.sort(myarr) # 순서대로 정렬
print("정렬:", sorted_arr)

# 짝수인 경우 가운데 두 수의 평균 
if size % 2==0 :
    p=(size //2) -1 #인덱스가 0부터 시작하므로 -1
    q = p+1 
    median = (sorted_arr[p] + sorted_arr[q])/2
else:
    p = size //2 
    median = sorted_arr[p]

print("중앙값:", median)
```

```python
# 5. 중앙값(Numpy 사용)
print( "중앙값:", np.median(myarr))
```

```python
# 6. 1사분위 수 
x = np.sort(myarr) # 데이터 정렬 
n = len(x) #총 도수 
q1_pos = (n-1)*0.25 # 위치 계산 (p)
i = int(np.floor(q1_pos)) #아래쪽 인덱스(i)
f = q1_pos - i #보간비율(f)

# 보간(interpolation)
Q1 = x[i] + f* (x[i+1] - x[i])
print("직접 계산한 1사분위 수 (Q1):", Q1)

# numpy 내장 함수와 비교
print("Numpy로 계산한 1사분위수:", np.quantile(myarr, 0.25))
```

```python
# 7. 3사분위 수 
x = np.sort(myarr) # 데이터 정렬 
n = len(x) # 총 도수
q3_pos = (n-1) *0.75 # 위치 계산 (p)
i = int(np.floor(q3_pos)) #아래쪽 인덱스(i) 
f = q3_pos -i # 보간비율(f)

# 보간 (interpolation)
Q3 = x[i] + f* (x[i+1] -x[i])
print("직접 계산한 3사분위수 (Q3):", Q3)

# numpy 내장 함수와 비교 
print("Numpy로 계산한 3사분위수:", np.quantile(myarr, 0.75))
```

## #04. 데이터의 범위, 산포를 나타내는 기초 통계량 

```python
# 1. 데이터 구간(Range)
print("데이터 구간:", np.max(myarr) - np.min(myarr))
```

```python
# 2. 사분위 수 구간 (IQR)
q3 = np.quantile(myarr, 0.75) 
q1 = np.quantile(myarr, 0.25)
print("사분위수 구간(IQR):", q3 - q1)
```

```python
# 3. 분산, 표준편차 
print("분산:", np.var(myarr))
print("표준편차:", np.std(myarr))
```

## #05. 2차원 배열의 기술 통계량 

```python
# 1. 각 열끼리 연산 (세로 연산)
mydata = np.array([[98, 72, 80, 64, 88], 
                   [65, 79, 59, 81, 74]])

s1 = np.sum(mydata, axis =0) 
print(type(s1))
s1
```

## # 1. 연습문제 - 게임 점수 데이터 분석 

```python
Alpha = np.array([88, 92, 76, 100, 95, 81, 79, 98, 85, 87])

print("도수: ", len(Alpha))
print("평균:", np.average(Alpha))
print("표준편차:", np.std(Alpha))
print("최소값:", np.min(Alpha))
print("Numpy로 계산한 1사분위수:", np.quantile(Alpha, 0.25))
print("중앙값: ",np.median(Alpha))
print("Numpy로 계산한 3사분위수:", np.quantile(Alpha, 0.75))
print("최대값:", np.max(Alpha))
```

```python
Bravo = np.array([65, 99, 78, 72, 80, 91, 75, 82, 89, 70])

print("도수: ", len(Bravo))
print("평균:", np.average(Bravo))
print("표준편차:", np.std(Bravo))
print("최소값:", np.min(Bravo))
print("Numpy로 계산한 1사분위수:", np.quantile(Bravo, 0.25))
print("중앙값: ",np.median(Bravo))
print("Numpy로 계산한 3사분위수:", np.quantile(Bravo, 0.75))
print("최대값:", np.max(Bravo))
```
