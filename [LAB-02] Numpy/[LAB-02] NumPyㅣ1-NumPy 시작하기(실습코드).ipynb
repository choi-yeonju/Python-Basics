## #03. 리스트와 동일한 특성

### 1. 배열의 크기

- `len()` 함수를 사용하여 배열의 크기를 확인할 수 있다.

```python
arr = np.array([10,30,50,70,90])
size = len(arr)
print("배열의 원소는 %d개 입니다." % size)
```

### 2.인덱싱과 슬라이싱

- 리스트와 동일한 방법으로 인덱싱과 슬라이싱이 가능하다.

```python
mydata = np.array([5, 10, 15, 20, 25])

# 인덱싱을 사용한 원소의 접근
print(mydata[0])
print(mydata[1])
print(mydata[3])

# 슬라이싱을 사용한 구간 추출
print(mydata[1:4])
print(mydata[-4:-1])
```

### 3. 반복문을 통한 활용

- 리스트와 동일하게 사용할 수 있다.

```python
mydata = np.array([100, 110, 120, 130, 140])

for i, v in enumerate(mydata):
    print("%d번째 원소 >> %d" % (i, v))
```

### 4. 논리연산 (2) - 조건식에 따른 원소 추출

```python
gt_or_lt = mypoint[np.logical_or(mypoint < 80, mypoint > 90)]
gt_or_lt
```

## #08. 조건 분기

### 1. `if ~ else` 효과

- `np.where()`은 주어진 조건에 따라 둘 중 하나의 값을 선택적으로 반환한다.
  - 조건 : `exam >= 80`
  - 조건이 참인 경우 : `'합격'`
  - 조건이 거짓인 경우 : `'불합격'`

```python
exam = np.array([60, 83, 75, 92, 86])

# 첫 번째 파라미터로 주어진 조건이 참일 경우 '합격',
# 그렇지 않을 경우 '불합격'
result = np.where(exam >= 80, '합격', '불합격')
result
```

### 2. 다중 조건 분기

```python
exam = np.array([60, 85, 51, 79, 94, 88])

# 학점을 부여하기 위한 점수의 구간을 설정하는 조건들을 리스트로 설정
conditions = [ (exam >= 90),   # if
               (exam >= 80),   # elif
               (exam >= 70)    # elif
]

# 조건에 따라 부여될 학점
value = ['A', 'B', 'C']

# 조건과 조건을 판단하기 위한 값을 지정함
# default는 조건에 맞는 값이 아닌 경우 사용됨 (생략 가능)
result = np.select(conditions, value, default='F')
result
```
