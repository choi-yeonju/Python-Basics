## #02. NumPy의 histogram : 도수 분포도

### 1. 5개의 구간으로 나누기

- `np.histogram(데이터, 구간수)` 함수를 사용하여    
  구간별 데이터수(hist)와 구간 범위(bins)를 리턴받는다.

```python
hist, bins = np.histogram(point, 5)
print("구간별 데이터 수:", hist)
print("구간 경계:", bins)
```

### 2. 도수 분포표 구성

- 구간별 데이터 수: [4 4 4 3 2]
- 구간 경계: [ 55.  64.  73.  82.  91. 100.]

```python
s = len(hist)
for i in range(0, s):
    if i + 1 < s:
        tpl = "%d이상 %d미만: %d개"
    else:
        tpl = "%d이상 %d이하: %d개"

    print(tpl % (bins[i], bins[i+1], hist[i]))
```

### 3. 데이터 구간을 직접 정의하기

- 구간을 지정한 리스트 객체를 histogram 함수에 두 번째 파라미터로 전달한다.

```python
hist, bins = np.histogram(point, [50, 60, 70, 80, 90, 100])
print("구간별 데이터 수:", hist)
print("구간 경계:", bins)
```

## #03. 도수의 종류

### 1. 절대 도수

- 어떠한 변수에 대한 실제 빈도수
- `np.histogram()`에서 리턴하는 `hist`값이 절대도수임.

### 2. 상대 도수

- 절대도수를 백분률로 환산한 값
- 각 도수를, 도수의 총 합으로 나눈 값

```python
relative_freq = hist / np.sum(hist)
relative_freq
```

### 3. 누적 도수

절대도수에서 자기 앞 도수를 모두 다 더한 값

```python
cum_freq = []

previous = 0

for i in hist:
    previous += i
    cum_freq.append(previous)

cum_freq
```

## 연습문제

### 연습문제 (1)

```python
import numpy as np

# 1. numpy 배열 생성
sales_data = np.array([18, 25, 33, 21, 28, 40, 15, 3, 9, 17, 22, 29, 38, 45, 11, 6, 31, 24, 27, 35, 12, 19, 39, 42, 7, 2, 20, 34, 26, 30])

# 2. 도수분포 분석
bins = [0, 10, 20, 30, 40, 50]
hist, bin_edges = np.histogram(sales_data, bins=bins)

# 상대도수 계산
relative_freq = hist / np.sum(hist)

# 누적도수 계산
cum_freq = []
previous = 0
for h in hist:
    previous += h
    cum_freq.append(previous)

# 3. 도수분포표 출력
print("--- 일일 판매량 분석 ---")

for i in range(len(hist)):
    # 구간의 마지막인지 확인하여 레이블 조정
    if i == len(hist) - 1:
        label = f"{bin_edges[i]} 이상 {bin_edges[i+1]} 이하"
    else:
        label = f"{bin_edges[i]} 이상 {bin_edges[i+1]} 미만"

    # 출력 템플릿
    tpl = f"구간: {label} | 절대도수: {hist[i]} | 상대도수: {relative_freq[i]:.3f} | 누적도수: {cum_freq[i]}"
    print(tpl)
```
