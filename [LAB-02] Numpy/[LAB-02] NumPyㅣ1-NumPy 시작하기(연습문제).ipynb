# [LAB-02] 1. NumPy 시작하기

## 연습문제

```python
import numpy as np

# 1. 배열 생성
arr = np.array([10, 20, 30, 40, 50])

# 2. 배열의 길이 출력
print("배열의 길이:", len(arr))

# 3. 인덱싱 / 슬라이싱
print("첫 번째 원소:", arr[0])
print("마지막 두 개 원소:", arr[-2:])

# 4. 반복문 + enumerate
for i, v in enumerate(arr):
    print("%d번째 원소 >> %d" % (i, v))
```
