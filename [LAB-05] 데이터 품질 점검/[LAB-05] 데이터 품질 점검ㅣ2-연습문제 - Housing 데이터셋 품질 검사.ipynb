### 질문 내용을 통해 도출한 결론

#### Step 1·2

- `price`와 `area`는 평균 > 중앙값, 왜도 > 1로 확인되어 **우편향 분포**이다.
- 따라서 두 변수는 **로그 변환 우선 대상**으로 식별된다.

#### Step 3·4

- 변동계수(CV)는 `area = 0.421`, `price = 0.392`로, **`area`의 상대 변동성이 약간 더 크다.**
- IQR 기준 상한 초과 비율은 `area 2.20%`, `price 2.75%`로 확인되었다.

#### Step 5

- 범주형 시설 변수는 가격 프리미엄과 관련이 있다.
  - `airconditioning = yes`는 `no` 대비 평균 가격이 **약 43.45% 높다.**
  - `prefarea = yes`는 `no` 대비 평균 가격이 **약 32.85% 높다.**
  - `hotwaterheating = yes`는 `no` 대비 평균 가격이 **약 17.58% 높다.**

- 즉, **시설 보유 여부와 선호지역 여부는 가격 프리미엄 형성에 유의미한 영향을 주는 변수**로 볼 수 있다.
